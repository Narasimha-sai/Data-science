```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier

root=Tk()
root.title("BMI CALCULATOR")
root.configure(background='RED')

#Function to predict the outcome
def predict():
    name=ent1.get()
    heightin=float(ent2.get())
    weight=int(ent3.get())
    gender=int(g.get())
    
    #Fetching thez Data
    data=pd.read_csv("BMI_Data.csv")
    data_dataframe=pd.DataFrame(data)
    data_dataframe
    # Spliting-out validation dataset
    array = data_dataframe.values
    X = array[:,0:4]
    Y = array[:,4]
    seed = 7
    X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=0.20, random_state=seed)
    # Predicting DataSet
    height=0.0254*heightin
    bmi=weight/(height**2)
    P=[[gender,weight,height,bmi]]
    
    # Decision Tree Model
    clf = DecisionTreeClassifier()
    clf = clf.fit(X_validation,Y_validation)
    result=str(clf.predict(P))
    messagebox.showinfo (title='YOUR BODY MASS INDEX', message="Hello " '%s \n As per BMI Report, \n You are %s' %(name,result))
    
#Adding the logo
imagelogo=PhotoImage(file='hearts.png')
logo = Label(root, image=imagelogo,background='RED')
logo.place(x=75, y=-20)
lable1 = Label(root, text="BODY MASS INDEX ",background='RED',fg="SKY BLUE",font = "Arial 80 bold").place(x=330, y=30)

#Adding image and text
imageex=PhotoImage(file='doc.png')
imgg = Label(root, image=imageex,background='RED',height=500,width=800)
imgg.place(x=740, y=230)
lable1 = Label(root, text="Health is the ability of a biological \nsystem to acquire,convert, allocate, distribute,\n and utilize energy with maximum efficiency.\n The World Health Organization (WHO) \n defined human health as a state of \n complete physical, mental and social well-being \n and not merely the absence of disease or infirmity.",background='LIGHT BLUE',fg="#D2691E",font = "Arial 12 bold",anchor="w", borderwidth=2,relief="solid",highlightcolor="#D2691E").place(x=20, y=350)

#Inserting the labels
lable1 = Label(root, text="Name:",background='RED',font = "Arial 35 bold",fg="LIGHT GREEN").place(x=480, y=250)
lable2 = Label(root, text="Gender:",background='RED',font = "Arial 35 bold",fg="LIGHT GREEN").place(x=475, y=350)
lable3 = Label(root, text="Height:",background='RED',font = "Arial 35 bold",fg="LIGHT GREEN").place(x=480, y=450)
lablex1 = Label(root, text="(in inches)",background='RED',font = "Arial 15 ",fg="LIGHT GREEN").place(x=480, y=508)
lable4 = Label(root, text="Weight:",background='RED',font = "Arial 35 bold",fg="LIGHT GREEN").place(x=480, y=570)
lableex2 = Label(root, text="(in kgs)",background='RED',font = "Arial 15",fg="LIGHT GREEN").place(x=480, y=630)

#Inputting the data
large_font = ('Verdana',15)
entry1Var = StringVar(value='Large Font!')
ent1 = Entry(root,font=large_font)
ent1.place(x=690, y=270)
ent2 = Entry(root,font=large_font)
ent2.place(x=690, y=470)
ent3 = Entry(root,font=large_font)
ent3.place(x=690, y=590)

g=IntVar()
radio1=Radiobutton(root,text="Male",variable=g,value=0,background='RED',font = "Arial 15",fg="GREEN")
radio1.place(x=700, y=365)
radio2 = Radiobutton(root, text="Female", variable=g, value=1,background='RED',font = "Arial 15",fg="GREEN")
radio2.place(x=850, y=365)

button1 = Button(root, text="Predict",command=predict,width=10,background='LIGHT BLUE',font = "Arial 15",fg="GREEN")
button1.place(x=700, y=700)
button2 = Button(root, text='Exit', command=root.destroy,width=10,background='LIGHT BLUE',font = "Arial 15",fg="GREEN")
button2.place(x=900,y=700)
root.mainloop()
```
